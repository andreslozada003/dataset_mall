```python
#2 Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario. 
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Dividir el dataset en características (X) y variable objetivo (y)
X = mall_data.drop(columns=['CustomerID', 'Gender'])
y = mall_data['Gender']
```

```python
# Columnas numéricas y categóricas
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns
```

```python
# Preprocesamiento de datos
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
# Aplicar preprocesamiento
X_processed = preprocessor.fit_transform(X)
```

```python
# Convertir X_processed a DataFrame para facilitar su manipulación
X_processed = pd.DataFrame(X_processed)
```

```python
# Ver los primeros registros del dataset preprocesado
print(X_processed.head())
```

```python
# Ver la forma del dataset preprocesado
print(X_processed.shape)
```

```python
#Seleccionar las características más relevantes para entrenar el modelo utilizando selección de características.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
# Entrenar un clasificador de bosques aleatorios para evaluar la importancia de las características
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_processed, y)
```

```python
# Seleccionar características con importancia por encima de un umbral
selector = SelectFromModel(rf, threshold='mean', prefit=True)
X_selected = selector.transform(X_processed)
```

```python
# Obtener el índice de las características seleccionadas
feature_idx = selector.get_support()
```

```python
# Obtener los nombres de las características seleccionadas
selected_features = X.columns[feature_idx]
```

```python
# Imprimir las características seleccionadas
print("Características seleccionadas:")
print(selected_features)
```

```python
#Aquí te muestro cómo entrenar ambos modelos con diferentes configuraciones de hiperparámetros

# K-means
```

```python
import os
import warnings
from sklearn.cluster import KMeans
```

```python
# Silenciar la advertencia sobre los núcleos físicos
os.environ['LOKY_MAX_CPU_COUNT'] = '8'  # Cambia '8' por el número de núcleos físicos de tu máquina

# Evitar la fuga de memoria con KMeans en Windows con MKL
os.environ['OMP_NUM_THREADS'] = '1'

# Configuración de hiperparámetros
k_values = [2, 3, 4, 5, 6]  # Número de clusters

# Entrenamiento y evaluación de modelos
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", category=UserWarning)
        kmeans.fit(X_processed)
    # Evaluar el modelo, por ejemplo, con inercia
    print(f"K-means con {k} clusters - Inercia: {kmeans.inertia_}")
```

```python
# Clustering Jerárquico
```

```python
from sklearn.cluster import AgglomerativeClustering

# Configuración de hiperparámetros
linkage_methods = ['ward', 'complete', 'average', 'single']  # Métodos de enlace

# Entrenamiento y evaluación de modelos
for method in linkage_methods:
    hierarchical = AgglomerativeClustering(n_clusters=5, linkage=method)
    hierarchical.fit(X_processed)
    # Evaluar el modelo, por ejemplo, con el número de clusters finales
    print(f"Clustering Jerárquico con método de enlace '{method}': {len(np.unique(hierarchical.labels_))} clusters")
```

```python
from sklearn.metrics import silhouette_score

# Calcular el coeficiente de Silhouette para K-means
silhouette_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_processed)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(X_processed, labels))

# Imprimir los resultados
print("Coeficientes de Silhouette para K-means:")
for k, score in zip(k_values, silhouette_scores):
    print(f"K-means con {k} clusters: {score}")
```
